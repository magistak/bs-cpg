# bs-cpg

> A python package for downloading and processing RRBS DNA methylation data.

**bs-cpg** is a Python library designed to streamline the acquisition and processing of Reduced Representation Bisulfite Sequencing (RRBS) DNA methylation data.

**Key Capabilities:**
* **Data Retrieval**: Automates downloading processed RRBS data (BED format) from GEO.
* **Reference Management**: Downloads and prepares bgzipped reference genomes and liftover chain files from UCSC.
* **Coordinate Processing**: Handles genomic liftover and validates CpG sites across different genome builds.

This is the main documentation homepage for bs-cpg. For comprehensive module-specific documentation, API references, and examples, see the sections below. For development guidance, see the [Developer Guide](04_developers.ipynb).

## Quick Start

### Basic Workflow Example

Here's a simple workflow to get started with bs-cpg:

```python
from bs_cpg.setup import read_sample_cpg, get_base_data_path
from bs_cpg.liftover_ps import cpg_reads, cpg_percent, liftover_positions

# 1. Load sample CpG data
df = read_sample_cpg(["chromosome", "pos"])

# 2. Validate CpG sites
reads = cpg_reads(df["chromosome"], df["pos"], "hg19", index_base=1)
percent_valid = cpg_percent(reads)
print(f"Valid CpG sites: {percent_valid:.1f}%")

# 3. Perform liftover to hg38
new_chroms, new_pos = liftover_positions(
    chromosomes=df["chromosome"],
    positions=df["pos"],
    genome_from="hg19",
    genome_to="hg38",
    index_base_from=1,
    index_base_to=1
)
```
